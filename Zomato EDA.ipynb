```python
df_country = pd.read_excel('Zomato dataset\Country-Code.xlsx')
df_country.head()
```

```python
final_df = pd.merge(df, df_country, on='Country Code',how='left')
```

```python
final_df.head(3)
```

```python
# To check data types
final_df.dtypes
```

```python
#To see how many records are in each country(General observation)
final_df.Country.value_counts()
```

```python
country_names = final_df.Country.value_counts().index
```

```python
country_val=final_df.Country.value_counts().values
```

```python
## Pie Chart - Top 3 countries that uses zomato
plt.pie(country_val[:3],labels=country_names[:3],autopct='%1.2f%%')
```

### Observation:
 Zomato maximum records or transaction are from India
 After that USA and then United Kingdoms

```python
final_df.columns
```

```python
## Want to see from which countries what and how much ratings are coming
ratings = final_df.groupby(['Aggregate rating','Rating color','Rating text']).size().reset_index().rename(columns={0:'Rating Count'})
```

```python
ratings
```

## Observation
1. When rating is between 4.5 to 4.9 ----> excellent
2. When rating is between 4.0 to 4.4 ----> very good
3. When rating is between 3.5 to 3.9 ----> good
4. When rating is between 3.0 to 3.4 ----> average
5. When rating is between 2.5 to 2.9 ----> average
6. When rating is between 2.0 to 2.4 ----> poor

```python
matplotlib.rcParams['figure.figsize']=(12,6) # For making the plot look bigger and clear
sns.barplot(x="Aggregate rating", y="Rating Count", data=ratings)
```

```python
sns.barplot(x="Aggregate rating", y="Rating Count", hue="Rating color", data=ratings)
```

```python
# palette is an attribute present in barplot where you can give your own colours
sns.barplot(x="Aggregate rating", y="Rating Count", hue="Rating color", data=ratings, palette=['black','red','orange','yellow','green','green'])
```

### Observation:
1. Not Rated count is very high
2. Maximum number of rating are between 2.5 to 3.4
